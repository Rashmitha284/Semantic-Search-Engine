```python
import re

def clean_text(text):
    # Remove timestamps
    text = re.sub(r'\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}\r\n', '', text)
    # Remove HTML tags
    text = re.sub(r'<[^>]*>', '', text)
    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    # Remove non-ASCII characters
    text = ''.join(char for char in text if ord(char) < 128)
    return text.strip()

# Apply the clean_text function to the 'content' column
df['decode_content'] = df['decode_content'].apply(clean_text)

# Print the DataFrame to verify the changes
print(df.head())
```

```python
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.stem import PorterStemmer

# Instantiate the Porter stemmer
stemmer = PorterStemmer()
```

```python
def preprocess(raw_text, flag):
    # Removing special characters and digits
    text = re.sub("[^a-zA-Z]", " ", raw_text)
    
    # change text to lower case
    text = text.lower()

    # tokenize into words
    words = word_tokenize(text)
    
    # remove stop words                
    words = [word for word in words if word not in stopwords.words("english")]
    
    # Stemming/Lemmatization
    if(flag == 'stem'):
        words = [stemmer.stem(word) for word in words]
    else:
        words = [lemmatizer.lemmatize(word) for word in words]

    preprocessed_text = " ".join(words)
    words_in_preprocessed_text = len(words)
    
    return pd.Series([preprocessed_text, words_in_preprocessed_text])
```

```python
lemmatizer = WordNetLemmatizer()

# Assuming 'file_content' is the name of the column containing the text
def lemmatize_text(text):
    if isinstance(text, str):
        # Tokenize the text into words
        words = word_tokenize(text)

        # Lemmatize each word
        lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

        # Join the lemmatized words back into a single string
        lemmatized_text = ' '.join(lemmatized_words)

        return lemmatized_text
    else:
        # If the input is not a string, print an error message
        print("Error: Input is not a string")
        return None

# Apply lemmatization to the entire column
df['decode_content'] = df['decode_content'].apply(lemmatize_text)

# Print the DataFrame to verify the changes
print(df.head())
```
